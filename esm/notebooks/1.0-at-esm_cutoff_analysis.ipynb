```python
dfs = []
for rfp in os.listdir(fp):
    if os.path.exists(fp + rfp + "/unfiltered_proteins.csv"):
        df = pd.read_csv(fp + rfp + "/unfiltered_proteins.csv")
        df["target"] = rfp.split("_")[0]
        dfs.append(df)
df = pd.concat(dfs)
```
